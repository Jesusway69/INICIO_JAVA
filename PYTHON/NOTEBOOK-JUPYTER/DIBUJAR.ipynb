```python
import numpy as np

# DEFINIR LOS DATOS QUE VOY A GRAFICAR

x = np.linspace(1,2*np.pi)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = 1 / np.sin(x)

plt.figure()

plt.subplot(231) 
plt.plot(x,y1,color='blue')

plt.subplot(232)
plt.plot(x,y2,color='red')

plt.subplot(233)
plt.plot(x,y3,color='red')

plt.subplot(234)
plt.plot(x,y4,color='red')

plt.subplot(235)
plt.plot(x,y2,color='red')

plt.subplot(236)
plt.plot(x,y2,color='red')


plt.show()
```
